# Loops

```python
# 2 Typres of - Loop helps to iterate over sequence or repeat actions.
# For loop & While loop

## for loop: (faster)

for i in range(5):
  print(i)
```

```python
## while loop

i = 0

while i < 5:
  print(i)
  i = i + 1
```

# Loop control statments

```python
# break : it exits the loop
# continious : it skips the current iteration

for i in range(10):
  if i == 5:
    break
  if i == 3:
    continue
  print(i)

```

```python
## question: write a for loop that counts the exact no of times a letter 'a' appers in a string aashish
count = 0
s = 'Aaansh'

for i in s:
  if i == 'a' or i == 'A':
    count = count + 1
print(count)
```
