```python
# Calculator - Addition, Subtraction, Multiplication, Division, Power, Square Root (optional), Logarithm, Trigonometric Functions (Sin, Cos, Tan)

num1 = int(input("Enter the first number: "))
num2 = int(input("Enter the second number: "))

function = input("Enter the operation you want to perform: ").lower()

if function == "add":
  print(num1 + num2)
elif function == "substract":
  print(num1-num2)
elif function == "multiply":
  print(num1*num2)
elif function == "divide":
  print(num1/num2)
elif function == "power":
  print(num1**num2)
elif function == "root":
  print(num1**0.5)
else:
  print("Invalid input")
```
