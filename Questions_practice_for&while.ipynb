```python
#15 Menu-Driven Program

print("Menue")
print("1. display no 1-10")
print("2. print even from 1-20")
print("3. Muultiplication table")
print("4. Exit")

while True:
  choice = int(input("enter what you want to carry out"))
  if choice == 1:
    for i in range(1,11):
      print(i)
  elif choice == 2:
    for i in range(20):
      if i % 2 == 0:
        print(i)
  elif choice == 3:
    no = int(input("Enter a number: "))
    for i in range(1,11):
      print(no, 'X', i, '=', no*i)
  else:
    print("Exited")
    break

```
